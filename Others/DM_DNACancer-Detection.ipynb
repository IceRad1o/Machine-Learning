```python
##---Extract data from xlsx file---##
import xlrd
file = xlrd.open_workbook(r'Dataset/data2.xlsx')
sheet = file.sheet_by_name('7')
data2 = []
for i in range(1, sheet.nrows):
    row_content = []
    for j in range(sheet.ncols):
        cell = sheet.cell_value(i, j)
        row_content.append(cell)
    data2.append(row_content)
```

```python
##---Extract data from csv file---##
import pandas as pd
df = pd.read_csv(r'Dataset/data1.csv')
data1=df.values.tolist()
```

```python

##---Combine the data from csv and xlsx---##
def GetIndex(list,value):#This is the function to obtain the index of a certain value in the give list
    for i in range(len(list)):
        for j in range(len(list[i])):
            if list[i][j] == value:
                index1 = [i,j]
    return index1

p = GetIndex(data2,data1[422][0])
print(p)
print(p[0])
```

```python

x=[]
y=[]
for i in range(len(data1)):
    p = GetIndex(data2,data1[i][0])
    data1[i].append(data2[p[0]][1])# add Median cfDNA Fragment Size (bp) from data2 to data1
    data1[i].append(data2[p[0]][2])# add Correlation of GC Corrected Fragment Ratio Profile to Median Fragment Ratio Profile of Healthy Individuals from data2 to data1
    if 'Healthy' in data1[i]:# Convert the character lable to a 0-1 lable
        data1[i][2] = 0
    else:
        data1[i][2] = 1

    x1=data1[i][:]
    del x1[0]
    del x1[1]
    x.append(x1)#feature
    y.append(data1[i][2])  # lable
```
